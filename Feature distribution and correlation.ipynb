```python
# Return unbiased standard deviation 

std=(comparison['AirbnbAccuracy']-comparison['PipeAccuracy']).std()
std.round(3)
```

```python
# Standard error of the mean

sem1=(comparison['AirbnbAccuracy']-comparison['PipeAccuracy']).sem()
sem2 = (comparison['AirbnbCleanliness']-comparison['PipeCleanliness']).sem()
sem3 =(comparison['AirbnbLocation']-comparison['PipeLocation']).sem()
sem4= (comparison['AirbnbCheck-in']-comparison['PipeCheckin']).sem()
sem5=(comparison['AirbnbValue']-comparison['PipeValue']).sem()
sem6=(comparison['AirbnbCommunication']-comparison['PipeCommunication']).sem()
sem={'Standard error of mean':[sem1,sem2,sem3,sem4,sem5,sem6]}
pd.DataFrame(sem, index=['Accuracy', 'Check-in','Cleanliness','Communication','Location','Value'])
```
